```python
import cv2
import os
import mediapipe as mp
mp_face_detection = mp.solutions.face_detection
LABELS = ["Con_cubrebocas", "Sin_cubrebocas"]
# Leer el modelo
face_mask = cv2.face.LBPHFaceRecognizer_create()
face_mask.read("CubrebocasModel.xml")
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
with mp_face_detection.FaceDetection(
     min_detection_confidence=0.5) as face_detection:
     while True:
          ret, frame = cap.read()
          if ret == False: break
          frame = cv2.flip(frame, 1)
          height, width, _ = frame.shape
          frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
          results = face_detection.process(frame_rgb)
          if results.detections is not None:
               for detection in results.detections:
                    xmin = int(detection.location_data.relative_bounding_box.xmin * width)
                    ymin = int(detection.location_data.relative_bounding_box.ymin * height)
                    w = int(detection.location_data.relative_bounding_box.width * width)
                    h = int(detection.location_data.relative_bounding_box.height * height)
                    if xmin < 0 and ymin < 0:
                         continue
                    #cv2.rectangle(frame, (xmin, ymin), (xmin + w, ymin + h), (0, 255, 0), 5)
                    face_image = frame[ymin : ymin + h, xmin : xmin + w]
                    face_image = cv2.cvtColor(face_image, cv2.COLOR_BGR2GRAY)
                    face_image = cv2.resize(face_image, (72, 72), interpolation=cv2.INTER_CUBIC)
                    
                    result = face_mask.predict(face_image)
                    #cv2.putText(frame, "{}".format(result), (xmin, ymin - 5), 1, 1.3, (210, 124, 176), 1, cv2.LINE_AA)
                    if result[1] < 150:
                         color = (255, 0, 0) if LABELS[result[0]] == "Con_cubrebocas" else (0, 0, 255)
                         cv2.putText(frame, "{}".format(LABELS[result[0]]), (xmin, ymin - 15), 2, 1, color, 1, cv2.LINE_AA)
                         cv2.rectangle(frame, (xmin, ymin), (xmin + w, ymin + h), color, 2)
          cv2.imshow("Frame", frame)
          k = cv2.waitKey(1)
          if k == 27:
               break
cap.release()
cv2.destroyAllWindows()
```

```python
# Utilizando cascade 

import cv2

import os
import cv2 as cv
import mediapipe as mp

cap = cv.VideoCapture(0)
mp_face_detection = mp.solutions.face_detection
cubrebocas = cv.CascadeClassifier('Mark3/classifier/cascade.xm')

with mp_face_detection.FaceDetection(min_detection_confidence = 0.5) as face_detection:
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv.flip(frame, 1)
        height, width, _ = frame.shape
        frame_rgb = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
        
        results = face_detection.process(frame_rgb)

        if results.detections is not None:
            for detection in results.detections:
                xmin = int(detection.location_data.relative_bounding_box.xmin * width)
                ymin = int(detection.location_data.relative_bounding_box.ymin * height)
                wmin = int(detection.location_data.relative_bounding_box.width * width)
                hmin = int(detection.location_data.relative_bounding_box.height * height)
                if not (xmin<0 and ymin<0):
                    try:
                        face_image = frame[ymin-10:ymin+hmin, xmin:xmin+wmin]
                        face_image = cv.cvtColor(face_image, cv.COLOR_BGR2GRAY)
                        face_image = cv.resize(face_image, (120, 110), interpolation=cv.INTER_CUBIC)
                        rostros = cubrebocas.detectMultiScale(face_image, 5, 15)
                        for(x, y, w, h) in rostros:
                            frame = cv.rectangle(frame, (xmin,ymin-10), (xmin+wmin, ymin+hmin), (0, 0, 255), 2)
                        cv.imshow('Recortada', face_image)
                    except cv.error as e:
                        print('error: ', e)
        else:
            cv.putText(frame, 'No hay rostros', (10, 20), 1, 2.3, (0, 0, 255), 1, cv.LINE_AA)
        cv.imshow('Original', frame)
        k = cv.waitKey(1)
        if k == 27:
            break

cap.release()
cv.destroyAllWindows()
```

```python
# Detectar rostros con la camaras usando un dataset en xml 

import numpy as np
import cv2 as cv
import math 

rostro = cv.CascadeClassifier('haarcascade_frontalface_alt.xml')
cap = cv.VideoCapture(0)
while True:
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    rostros = rostro.detectMultiScale(gray, 1.3, 5)
    for(x, y, w, h) in rostros:
        frame = cv.rectangle(frame, (x,y), (x+w, y+h), (255, 0, 0), 5)
    cv.imshow('rostros', frame)
    k = cv.waitKey(1)
    if k == 27:
        break
cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv

# Cargar el clasificador de cascada Haar previamente entrenado
rostro = cv.CascadeClassifier('Mark/classifier/cascade.xml')

# Cargar la imagen esttica
imagen_path = 'p1.jpeg'  # Reemplaza con la ruta de tu imagen
imagen = cv.imread(imagen_path)

# Convertir la imagen a escala de grises
gray = cv.cvtColor(imagen, cv.COLOR_BGR2GRAY)

# Realizar la deteccin de rostros
rostros = rostro.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=15, minSize=(50, 50), maxSize=(100, 100))

# Dibujar rectngulos alrededor de los rostros detectados
for (x, y, w, h) in rostros:
    imagen = cv.rectangle(imagen, (x, y), (x + w, y + h), (0, 255, 0), 5)

# Mostrar la imagen resultante
cv.imshow('Deteccin de Rostros', imagen)
cv.waitKey(0)
cv.destroyAllWindows()
```
