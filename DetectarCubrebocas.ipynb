```python
import os
import cv2 as cv
import mediapipe as mp

mp_face_detection = mp.solutions.face_detection

face_mask = cv.face.LBPHFaceRecognizer_create()
face_mask.read('Mark3/classifier/cascade.xml')

cap = cv.VideoCapture(0)

with mp_face_detection.FaceDetection(min_detection_confidence = 0.5) as face_detection:
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv.flip(frame, 1)
        height, width, _ = frame.shape
        frame_rgb = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
        
        results = face_detection.process(frame_rgb)

        if results.detections is not None:
            for detection in results.detections:
                xmin = int(detection.location_data.relative_bounding_box.xmin * width)
                ymin = int(detection.location_data.relative_bounding_box.ymin * height)
                w = int(detection.location_data.relative_bounding_box.width * width)
                h = int(detection.location_data.relative_bounding_box.height * height)
                if not (xmin<0 and ymin<0):
                    try:
                        face_image = frame[ymin-100:ymin+h+100, xmin-100:xmin+w+100]
                        face_image = cv.cvtColor(face_image, cv.COLOR_BGR2GRAY)
                        face_image = cv.resize(face_image, (150, 150), interpolation=cv.INTER_CUBIC)
                        
                        result = face_mask.predict(face_image)
                        if(result[0]==1 and result[1]>85):
                            cv.putText(frame, "Con cubrebocas", (xmin, ymin-5), 1, 1.3, (0,255,0), 1, cv.LINE_AA)
                        else:
                            cv.putText(frame, "Sin cubrebocas", (xmin, ymin-5), 1, 1.3, (0,0,255), 1, cv.LINE_AA)
                    except cv.error as e:
                        print('error: ', e)
        cv.imshow('Original', frame)
        k = cv.waitKey(1)
        if k == 27:
            break

cap.release()
cv.destroyAllWindows()
```

```python
import os
import cv2 as cv
import mediapipe as mp

# Load the face detection model from mediapipe
mp_face_detection = mp.solutions.face_detection
face_detection = mp_face_detection.FaceDetection()

# Use cv2.face.LBPHFaceRecognizer_create() for OpenCV 3.x
# For OpenCV 4.x and later, use cv2.face.LBPHFaceRecognizer_create()
# or cv2.face.LBPHFaceRecognizer_create() directly
if cv.__version__.startswith('3.'):
    face_mask = cv.face.LBPHFaceRecognizer_create()
else:
    face_mask = cv.face.LBPHFaceRecognizer_create()

# Load the trained face recognition model
face_mask.read('Mark3/classifier/cascade.xml')

# Open the video capture
cap = cv.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture image")
        break

    # Convert the frame to grayscale for face recognition
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    # Perform face detection using mediapipe
    results = face_detection.process(gray)
    
    if results.detections:
        for detection in results.detections:
            bboxC = detection.location_data.relative_bounding_box
            ih, iw, _ = frame.shape
            bbox = int(bboxC.xmin * iw), int(bboxC.ymin * ih), \
                   int(bboxC.width * iw), int(bboxC.height * ih)
            
            # Extract the region of interest for face recognition
            face_roi = gray[bbox[1]:bbox[1] + bbox[3], bbox[0]:bbox[0] + bbox[2]]

            # Perform face recognition using LBPHFaceRecognizer
            label, confidence = face_mask.predict(face_roi)
            
            # Display the label and confidence on the frame
            cv.putText(frame, f'Label: {label}', (bbox[0], bbox[1] - 10),
                       cv.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            cv.putText(frame, f'Confidence: {confidence}', (bbox[0], bbox[1] - 40),
                       cv.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

            # Draw the bounding box around the face
            cv.rectangle(frame, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]),
                         (0, 255, 0), 2)

    # Display the frame
    cv.imshow('Frame', frame)

    # Exit the loop when 'q' is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv.destroyAllWindows()

```

```python
# Escala de grises 


import cv2

cap = cv2.VideoCapture(0)
faceClassif = cv2.CascadeClassifier('Mark6/classifier/cascade.xml')

min_size = (50, 50)

while True:
    ret, frame = cap.read()
    
    # Convertir la imagen a escala de grises
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detección de caras en la imagen en escala de grises
    faces = faceClassif.detectMultiScale(
        gray, 
        scaleFactor=1.3, 
        minNeighbors=60,
        minSize = min_size,
    )

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (159, 248, 227), 2)

    frame = cv2.flip(frame, 1)
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# CAPTURA DE CAMARA DE VIDEO Y DETECCIÓN DE ROSTRO CON HAARCASCADES

# IMPORTS PARA EL CÓDIGO
import cv2 as cv

# IMPORTS PARA DE HAARCASCADES
Rostro = cv.CascadeClassifier('Mark6/classifier/cascade.xml')

# OBTENCIÓN DE LA CAPTURA DE VIDEO
Cap = cv.VideoCapture(0)

# VARIABLE DEL TAMAÑO DE IMAGEN
min_size = (50, 50)
max_size = (50, 50)

# CICLO PARA DETECTAR ROSTRO Y DIBUJAR RECTANGULOS SOBRE EL MISMO
while True:
    # OBTENER IMAGEN A PARTIR DE LA CAPTURA DE VIDEO
    ret, img = Cap.read()
    
    # CONVERTIR LA IMAGEN CAPTURADA A ESCALA DE GRISES
    Gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    
    # VALORES PARA LA DETECCIÓN DEL ROSTRO
    Rostros = Rostro.detectMultiScale(
        Gris,
        scaleFactor = 1.3, # PARÁMETRO QUE ESPECIFICA CUÁNTO SE REDUCE EL TAMAÑO DE LA IMAGEN EN CADA ESCALA DE IMAGEN.
        minNeighbors = 10, # PARÁMETRO QUE ESPECIFICA CUÁNTOS VECINOS DEBE TENER CADA RECTÁNGULO CANDIDATO PARA RETENERLO.
        minSize = min_size, # TAMAÑO MÍNIMO POSIBLE DEL OBJETO. LOS OBJETOS MÁS PEQUEÑOS SE IGNORAN.
        maxSize = max_size # TAMAÑO MÁXIMO DE OBJETO POSIBLE. LOS OBJETOS MÁS GRANDES SE IGNORAN. SI MAXSIZE == MINSIZE EL MODELO SE EVALÚA EN UNA SOLA ESCALA.
    )
    
    # CICLO FOR PARA EL DIBUJADO DEL RECTANGULO SOBRE EL ROSTRO
    for(x, y, w, h) in Rostros:
        img = cv.rectangle(img, (x, y), (x+w, y+h), (234, 23, 23), 2)
        #img = cv.rectangle(img, (x, int(y+ h/2)), (x+w, y+h), (0, 255, 0), 2)
    
    # CREACIÓN DE VENTANA PARA LA CAPTURA DE VIDEO DENTRO DEL CICLO
    cv.imshow('Captura de video original', img)
    
    # TECLA "Q" PARA ROMPER EL CICLO
    if cv.waitKey(1) == ord('q'):
        break

# CIERRE DE VENTANAS
Cap.release
cv.destroyAllWindows()
```
